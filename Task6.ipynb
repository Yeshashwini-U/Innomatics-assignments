```python
#problem 1: Smart Parking Lot Management System

#function to manage parking lot
def manage_parking(capacity, vehicle_logs):
    #variable to store currently parked vehicles
    parked_vehicles = 0
    #loop through each vehicle log
    for log in vehicle_logs:
        #if vehicle enters, increase count
        if log == "IN":
            parked_vehicles += 1
        #if vehicle exits, decrease count
        elif log == "OUT":
            parked_vehicles -= 1
    #check parking status
    if parked_vehicles > capacity:
        status = "Parking Full"
    else:
        status = "Available"
    #display the result
    print("Currently Parked Vehicles:", parked_vehicles)
    print("Parking Status:", status)

#sample input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

#call the function
manage_parking(parking_capacity, vehicle_logs)
```

```python
#problem 2: Online Food Delivery Time Estimator

#function to estimate delivery time
def estimate_delivery_time(distance, traffic_level, weather):
    #base time calculation
    delivery_time = distance * 5
    #add delay based on traffic level
    if traffic_level == "High":
        delivery_time += 10
    elif traffic_level == "Medium":
        delivery_time += 5
    #add delay based on weather condition
    if weather == "Rainy":
        delivery_time += 5
    elif weather == "Stormy":
        delivery_time += 15
    #display the final estimated time
    print("Estimated Delivery Time:", delivery_time, "minutes")

#sample input
distance_km = 8
traffic = "High"
weather_condition = "Rainy"

#call the function
estimate_delivery_time(distance_km, traffic, weather_condition)
```

```python
#problem 3: Movie Theatre Seat Occupancy Analyzer

#function to analyze seat occupancy
def analyze_seat_occupancy(total_seats, booked_seats):
    #variable to count number of booked seats
    booked_count = 0
    #loop through the booked seats list
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1
    #calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    #determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    #display the result
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)

#sample input
total_seats = 200
booked_seats = [1] * 150

#call the function
analyze_seat_occupancy(total_seats, booked_seats)
```

```python
#problem 4: Cloud Server Load Classification System

#function to classify server load
def classify_server_load(cpu_readings):
    #variable to store total CPU usage
    total_cpu = 0
    #loop through the CPU readings list
    for reading in cpu_readings:
        total_cpu += reading
    #calculate average CPU usage
    average_cpu = total_cpu / len(cpu_readings)
    #classify server status based
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    #display the result
    print("Average CPU Load:", int(average_cpu), "%")
    print("Server Status:", status)

#sample input
cpu_data = [45, 60, 70, 85, 90]

#call the function
classify_server_load(cpu_data)
```

```python
#problem 5: Smart Classroom Resource Usage Monitor

#function to check resource usage
def monitor_resource_usage(resource_usage):
    #list to store overused resources
    overused = []
    #loop through the dictionary
    for resource, hours in resource_usage.items():
        #if usage is more than 8 hours, consider it overused
        if hours > 8:
            overused.append(resource)
    #display overused resources
    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

#sample input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

#call the function
monitor_resource_usage(usage)
```

```python
#problem 6: Online Event Registration Capacity Controller

#function to manage event registrations
def manage_event_registration(capacity, registrations):
    #variable to count confirmed registrations
    confirmed = 0
    #variable to count waitlisted users
    waitlisted = 0
    #loop through number of registrations
    for i in range(registrations):
        #if capacity is not full, confirm registration
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    #check registration status
    if waitlisted > 0:
        status = "Closed"
    else:
        status = "Open"
    #display the result
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

#sample input
event_capacity = 100
total_registrations = 105

#call the function
manage_event_registration(event_capacity, total_registrations)
```
