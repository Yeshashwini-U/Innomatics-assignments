```python
#problem 1: User Login Check

#given credentials
correct_username = "admin"
correct_password = "1234"
#taking input from user
username = input("Enter username: ")
password = input("Enter password: ")
#checking if both username and password match
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#problem 2: Pass / Fail Analyzer

#list of student marks
marks = [45, 78, 90, 33, 60]
#variables to count pass and fail students
pass_count = 0
fail_count = 0
#checking each student's marks
for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
#printing the final results
print("Total number of students:", len(marks))
print("Number of students who passed:", pass_count)
print("Number of students who failed:", fail_count)
```

```python
#problem 3: Simple Data Cleaner

#original list of names
names = [" Alice ", "bob", " CHARLIE "]
#empty list to store cleaned names
cleaned_names = []
#processing each name one by one
for name in names:
    #remove extra spaces
    name_without_spaces = name.strip()
    #convert to lowercase
    name_lowercase = name_without_spaces.lower()
    #add cleaned name to new list
    cleaned_names.append(name_lowercase)
#print the cleaned list
print("Original names:", names)
print("Cleaned names:", cleaned_names)
```

```python
#problem 4: Message Length Analyzer

#list of messages
messages = ["Hi", "Welcome to the platform", "OK"]
#analyzing each message
for message in messages:
    #calculate the length of the message
    message_length = len(message)
    #print the message and its length
    print(f"Message: '{message}' - Length: {message_length}")
    #check if message is longer than 10 characters
    if message_length > 10:
        print("  --> This message is too long!")
    print()
```

```python
#problem 5: Error Message Detector

#system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
#variable to count errors
error_count = 0
#processing each log entry
for log in logs:
    #check if the log entry is an ERROR
    if log == "ERROR":
        error_count = error_count + 1
#print the total error count
print("Total number of ERROR entries:", error_count)
```
