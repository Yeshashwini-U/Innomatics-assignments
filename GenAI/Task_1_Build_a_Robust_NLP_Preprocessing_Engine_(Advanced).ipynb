Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as two different words because of case sensitivity. "Love" with a capital L may appear at the beginning of a sentence or can be used as a proper noun (like in a name or title), while "love" is generally used as a common noun or verb. If we don’t convert text to lowercase during preprocessing, the model might think they are completely different tokens, which can increase vocabulary size unnecessarily and reduce model performance. That is why most NLP pipelines convert all text to lowercase to maintain consistency.

2. What happens if stopwords are not removed?

If stopwords (like "is", "the", "and", "in") are not removed, the dataset will contain many common words that do not add much meaning to the sentence. This can make the model focus on irrelevant information and increase processing time. It also increases the size of the feature space, which can slow down training and reduce accuracy in some cases. However, sometimes keeping stopwords is useful depending on the task.

3. Provide two real-world scenarios where removing stopwords can be harmful.

Sentiment Analysis: In sentences like "I do not like this product", the word "not" is very important. If we remove stopwords, it may become "I like this product", which changes the meaning completely.
Question Answering Systems: In questions like "What is the time now?", removing stopwords may remove words like "what" and "is", which are important to understand the intent of the question.

4. Explain the difference between stemming and lemmatization.
Stemming and lemmatization are both used to reduce words to their base form, but they work differently.

Stemming is a simple and fast process that cuts off prefixes or suffixes from a word. For example, "running", "runner", and "runs" may all become "run" or sometimes even "runn". It does not always produce a correct English word.
Lemmatization is more advanced and uses vocabulary and grammar rules to convert a word into its proper base form. For example, "running" becomes "run", and "better" becomes "good". It gives more accurate results but takes more time compared to stemming.

Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    # 1. convert to lowercase
    text = text.lower()

    # 2. remove urls (http, https, www)
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. handle repeated characters (like sooo -> so)
    # this keeps only 2 same characters max
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. remove special characters (keeping only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. remove very short words (length <=2), but keep "no", "not"
    words = text.split()
    cleaned_words = []

    for word in words:
        if len(word) > 2 or word in ['no', 'not']:
            cleaned_words.append(word)

    # join back into sentence
    text = " ".join(cleaned_words)

    return text


# Example test
sample = "WOW!!! This is soooo goooood!!! Visit http://example.com now or mail me at test@gmail.com. I have 2 dogs!!"
print(preprocess_text(sample))
```

Task 4: Token Analytics


```python
# function to calculate token stats
def token_analysis(sentence):
    tokens = sentence.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    # calculate average token length
    if total_tokens > 0:
        total_length = 0
        for word in tokens:
            total_length += len(word)
        avg_length = total_length / total_tokens
    else:
        avg_length = 0

    return total_tokens, unique_tokens, avg_length


# running analysis on same test sentences
for i, sentence in enumerate(test_sentences):
    cleaned = preprocess_text(sentence)

    total, unique, avg_len = token_analysis(cleaned)

    print("Example", i+1)
    print("Cleaned Sentence:", cleaned)
    print("Total Tokens    :", total)
    print("Unique Tokens   :", unique)
    print("Avg Token Length:", round(avg_len, 2))
    print("-" * 50)
```

Task 5: Frequency Analysis

```python
from collections import Counter

# collect all tokens from all cleaned sentences
all_tokens = []

for sentence in test_sentences:
    cleaned = preprocess_text(sentence)
    tokens = cleaned.split()

    # add tokens to list
    for word in tokens:
        all_tokens.append(word)

# create frequency counter
freq = Counter(all_tokens)

# top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in freq.most_common(10):
    print(word, ":", count)

print("-" * 40)

# top 5 least frequent words
# (convert to list and sort manually for beginner understanding)
least_common = sorted(freq.items(), key=lambda x: x[1])

print("Top 5 Least Frequent Words:")
for word, count in least_common[:5]:
    print(word, ":", count)
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    # loop through each sentence
    for text in text_list:
        # preprocess each sentence
        cleaned = preprocess_text(text)

        # store cleaned sentence
        clean_sentences.append(cleaned)

        # split into tokens
        tokens = cleaned.split()

        # add tokens to global list
        for word in tokens:
            all_tokens.append(word)

    # return result in dictionary format
    result = {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

    return result


# testing the full pipeline
output = full_pipeline(test_sentences)

print("All Tokens:")
print(output["tokens"])

print("\nClean Sentences:")
for sent in output["clean_sentences"]:
    print(sent)
```

Task 7: Error Handling

```python
import re

def preprocess_text(text):
    # check if input is empty or None
    if text is None or text.strip() == "":
        return ""

    # convert to lowercase
    text = text.lower()

    # remove urls
    text = re.sub(r'http\S+|www\S+', '', text)

    # remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # remove numbers
    text = re.sub(r'\d+', '', text)

    # handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # remove special characters (this also removes emojis)
    text = re.sub(r'[^a-z\s]', '', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # remove short words (<=2), keep "no", "not"
    words = text.split()
    cleaned_words = []

    for word in words:
        if len(word) > 2 or word in ['no', 'not']:
            cleaned_words.append(word)

    return " ".join(cleaned_words)



edge_cases = [
    "",                 # empty string
    "😍🔥😂",           # only emojis
    "1234567890",       # only numbers
]

for i, text in enumerate(edge_cases):
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    print("Case", i+1)
    print("Original:", text)
    print("Cleaned :", cleaned)
    print("Tokens  :", tokens)
    print("-" * 40)
```
