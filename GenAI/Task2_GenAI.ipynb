```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
pipe = pipeline("text-generation", model="google/flan-t5-base")

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
response = llm.invoke("Explain Artificial Intelligence in simple terms")
print(response)
```

```python
prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

formatted_prompt = prompt.format(topic="Machine Learning")
print(formatted_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt)

result = chain.invoke({"topic": "LangChain"})
print(result["text"])
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()

memory.save_context({"input": "Hi"}, {"output": "Hello!"})
memory.save_context({"input": "What is AI?"}, {"output": "Artificial Intelligence"})

print(memory.load_memory_variables({}))
```

```python
from langchain.chains import ConversationChain

conversation = ConversationChain(
    llm=llm,
    memory=ConversationBufferMemory()
)

print(conversation.predict(input="My name is Alex"))
print(conversation.predict(input="What is my name?"))
```

```python
from langchain.tools import Tool

def calculator(expression: str) -> str:
    return str(eval(expression))

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Useful for solving math problems"
    )
]
```

```python
from langchain.agents import initialize_agent

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True,
    max_iterations=1   # ✅ stops infinite loop
)

response = agent.invoke("What is 10 * 5?")
print(response)
```

```python
teaching_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to a {audience} in a {tone} tone"
)

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

def generate_prompt(topic, audience, tone):
    audience, tone = validate_inputs(audience, tone)
    return teaching_prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("LangChain", "beginner", "casual"))
```
