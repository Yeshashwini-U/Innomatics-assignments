```python
from langchain_core.prompts import ChatPromptTemplate

# Function to generate chat prompt based on role
def get_chat_prompt(role, topic):

    system_message = f"You are a {role}."
    user_message = "Explain {topic}"

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", user_message)
    ])

    return chat_template.format(topic=topic)

# Test cases
print(get_chat_prompt("teacher", "Neural Networks"))
print("\n-----------------\n")
print(get_chat_prompt("interviewer", "Neural Networks"))
print("\n-----------------\n")
print(get_chat_prompt("motivator", "Neural Networks"))
```

```python
# Validation Function
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validate audience
    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    # Validate tone
    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone


# Test cases
print(validate_inputs("beginner", "fun"))        # valid
print(validate_inputs("kids", "friendly"))      # invalid → should default
```

```python
from langchain_core.prompts import PromptTemplate

# Prompt Generator Function
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Define templates for each style
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    # Step 3: Select template
    if style not in templates:
        print("Invalid style! Defaulting to 'teaching'")
        style = "teaching"

    selected_template = templates[style]

    # Step 4: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    # Step 5: Generate final prompt
    return prompt.format(topic=topic, audience=audience, tone=tone)


# Test Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
from langchain_core.prompts import PromptTemplate

# One reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

reuse_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# 5 different inputs
test_inputs = [
    {"topic": "AI", "audience": "beginner", "tone": "fun"},
    {"topic": "Python", "audience": "intermediate", "tone": "formal"},
    {"topic": "Blockchain", "audience": "expert", "tone": "casual"},
    {"topic": "Cybersecurity", "audience": "beginner", "tone": "formal"},
    {"topic": "Data Science", "audience": "intermediate", "tone": "fun"}
]

# Run all
for inp in test_inputs:
    print(reuse_prompt.format(**inp))
```
