```python
#problem 1: Employee Performance Bonus Eligibility

#dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
#find the highest performance score
highest_score = max(employees.values())
#find all employees who have the highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
#join the names with comma for display
names_output = ", ".join(top_performers)
#print the result
print(f"Top Performers Eligible for Bonus: {names_output} (Score: {highest_score})")
```

```python
#problem 2: Search Query Keyword Analysis

#input search query
query = "Buy mobile phone buy phone online"
#convert the sentence to lowercase
query = query.lower()
#remove common punctuation
for ch in ".,!?":
    query = query.replace(ch, "")
#split the sentence into words
words = query.split()
#dictionary to store word frequency
word_count = {}
#count how many times each word appears
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
#dictionary to store words that appear more than once
result = {}
for word, count in word_count.items():
    if count > 1:
        result[word] = count
#display the result
print(result)
```

```python
#problem 3: Sensor Data Validation

#list of sensor readings where index = hour and value = reading
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
#list to store valid readings
valid_readings = []
#check each reading
for index, value in enumerate(sensor_readings):
    #even numbers are considered valid
    if value % 2 == 0:
        valid_readings.append((index, value))
#display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#problem 4: Email Domain Usage Analysis

#list of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
#dictionary to store count of each domain
domain_count = {}
#total number of emails
total_emails = len(emails)
#extract domain and count occurrences
for email in emails:
    #split the email to get the domain part
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
#calculate and print percentage usage of each domain
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#problem 5: Sales Spike Detection

#list of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]
#calculate average sales
total_sales = sum(sales)
average_sales = total_sales / len(sales)
#value 30% above averageis considered a spike
threshold = average_sales * 1.3
#check each day's sales
for i in range(len(sales)):
    if sales[i] > threshold:
        # Day number is index + 1
        print(f"Day {i+1}: {sales[i]}")
```

```python
#problem 6: Duplicate User ID Detection

#list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
#dictionary to count how many times each ID appears
id_count = {}
#count occurrences of each user ID
for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1
    else:
        id_count[uid] = 1
#display only duplicate IDs
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
