```python
#problem 1: Social Media – Post Engagement Analyzer

#function to analyze post engagement based on likes
def analyze_post_engagement(likes_list):
    #variable to store total number of likes
    total_likes = 0
    #loop through the list
    for like in likes_list:
        total_likes += like
    #check post status based on total likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    #display the result
    print("Total Likes:", total_likes)
    print("Post Status:", status)
#example list of likes
likes = [200, 150, 300, 400]
#call the function
analyze_post_engagement(likes)
```

```python
#problem 2: Healthcare – Medicine Stock Alert System

#function to check medicine stock level
def check_medicine_stock(stock):
    #check stock condition
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    #display the result
    print("Medicine Stock:", stock)
    print("Status:", status)
#example stock value
medicine_stock = 6
#call the function
check_medicine_stock(medicine_stock)
```

```python
#problem 3: Agriculture – Rainfall Adequacy Checker

#function to check whether rainfall is adequate or not
def check_rainfall(rainfall_list, required_level):
    total_rainfall = 0
    #loop to calculate total rainfall
    for value in rainfall_list:
        total_rainfall += value
    #calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)
    #check rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    #display the result
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)
#example rainfall data
rainfall_data = [60, 80, 75, 70, 75]
#required rainfall level
required = 70
#call the function
check_rainfall(rainfall_data, required)
```

```python
#problem 4: Social Media – Duplicate Account Detection

#function to check for duplicate usernames
def check_duplicate_accounts(usernames):
    #convert list to set
    unique_usernames = set(usernames)
    #if length is different, duplicates are present
    if len(usernames) != len(unique_usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
#example list of usernames
user_list = ["john123", "amy45", "john123", "raj99"]
#call the function
check_duplicate_accounts(user_list)
```

```python
#problem 5: Healthcare – Appointment Eligibility Checker

#function to check appointment eligibility
def check_appointment_eligibility(age):
    #check if patient is eligible
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    #display the result
    print("Patient Age:", age)
    print("Eligibility Status:", status)
#example patient age
patient_age = 21
#callthe function
check_appointment_eligibility(patient_age)
```

```python
#problem 6: Agriculture – Premium Crop Price Filter

#function to filter premium crop prices
def filter_premium_crops(price_list):
    premium_crops = []
    #loop through each price
    for price in price_list:
        #check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)
    #display the result
    print("Premium Crops:", premium_crops)
#example list of crop prices
crop_prices = [1500, 2500, 1800, 3200, 2000]
#call the function
filter_premium_crops(crop_prices)
```

```python
#problem 7: System Monitoring – Application Health Checker

#function to check application health
def check_application_health(error_count):
    #check system status based on number of errors
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    #display the result
    print("Error Count:", error_count)
    print("System Status:", status)
#example error count
errors = 7
#call the function
check_application_health(errors)
```

```python
#problem 8: Banking – Daily Transaction Limit Checker

#function to check daily transaction limit
def check_transaction_limit(amount):
    daily_limit = 50000
    #check if transaction is within the limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    #display the result
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
#example transaction amount
transaction_amount = 60000
#call the function
check_transaction_limit(transaction_amount)
```

```python
#problem 9: E-Learning – Student Attendance Eligibility System

#function to calculate attendance percentage and check exam eligibility
def check_exam_eligibility(attendance_list):
    present_days = 0
    #loop through the attendance list
    for day in attendance_list:
        if day == "P":
            present_days += 1
    total_days = len(attendance_list)
    #calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    #check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    #display the result
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)
#example attendance data
attendance = ["P", "P", "A", "P", "P"]
#call the function
check_exam_eligibility(attendance)
```

```python
#problem 10: Smart Electricity Bill Analyzer

#function to calculate electricity bill and usage level
def calculate_electricity_bill(units):
    bill = 0
    #calculate bill based on unit slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    #classify usage based on total bill
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage
#example units consumed
units_consumed = 250
#call the function
total_bill, usage_status = calculate_electricity_bill(units_consumed)
#display the result
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
