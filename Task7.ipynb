```python
#problem 1: Smart Light Controller

#create a class for Smart Light
class SmartLight:
    #constructor to initialize light name and default status
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    #method to turn the light ON
    def turn_on(self):
        self.status = "ON"
    #method to turn the light OFF
    def turn_off(self):
        self.status = "OFF"
    #method to display the current status of the light
    def display_status(self):
        print(self.light_name + " is " + self.status)

#take input from user
light_name = input("Light Name: ")
action = input("Action: ")
#create object of SmartLight class
my_light = SmartLight(light_name)
#check action and perform operation
if action == "ON":
    my_light.turn_on()
elif action == "OFF":
    my_light.turn_off()
else:
    print("Invalid Action")
#display the result
my_light.display_status()
```

```python
#problem 2: Employee ID Card System

#create a class to store employee details
class Employee:
    #constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    #method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

#take input from user
emp_name = input("Employee Name: ")
emp_id = input("Employee ID: ")
emp_department = input("Department: ")
#create object of Employee class
employee1 = Employee(emp_name, emp_id, emp_department)

#display the ID card details
employee1.display_id_card()
```

```python
#problem 3: Mobile Contact Record

#create a class to store contact details
class Contact:
    #constructor to initialize contact name and phone number
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
    #method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone_number)

#take input from user
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")
#create object of Contact class
contact1 = Contact(contact_name, phone_number)
#display contact details
contact1.display_contact()
```

```python
#problem 4: Product Price Tag Generator

#create a class to store product details
class Product:
    #constructor to initialize product name and price
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    #method to print formatted price tag
    def print_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))

#take input from user
product_name = input("Product Name: ")
price = input("Price: ")
#create object of Product class
product1 = Product(product_name, price)
#print the price tag
product1.print_price_tag()
```

```python
#problem 5: Movie Rating Display System

#create a class to store movie details
class Movie:
    #constructor to initialize movie name and rating
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    #method to display movie details
    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")

#take input from user
movie_name = input("Movie Name: ")
rating = input("Rating: ")

#create object of Movie class
movie1 = Movie(movie_name, rating)

#display movie information
movie1.display_movie()
```

```python
#problem 6: Delivery Address Manager

#create a class to store delivery details
class Delivery:
    #constructor to initialize customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    #method to print delivery details
    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

#take input from user
customer_name = input("Customer Name: ")
address = input("Address: ")

#create object of Delivery class
delivery1 = Delivery(customer_name, address)

#print delivery information
delivery1.print_delivery_details()
```
