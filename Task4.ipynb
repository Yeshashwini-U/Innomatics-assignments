```python
#problem 1: Mobile Recharge Validation System

#function to validate the recharge amount
def validate_recharge():
    #list of valid recharge plans
    valid_plans = [199, 299, 399, 599]
    while True:
        #taking recharge amount as input from user
        amount = int(input("Enter recharge amount: ₹"))
        #check if amount is less than 50
        if amount < 50:
            print("Recharge amount must be at least ₹50. Please try again.\n")
            continue
        #check if entered amount is in the list of valid plans
        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid recharge plan. Please enter a valid plan.\n")
#call the function
validate_recharge()
```

```python
#problem 2: Inventory Reorder Alert System

#function to check stock levels of products
def check_inventory_stock(products):
    #loop through each product
    for product, quantity in products.items():
        #check if stock is less than 15
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")
#dictionary containing product names and their stock quantities
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 18,
    "Wheat": 5
}
#call the function
check_inventory_stock(inventory)
```

```python
#problem 3: Student Result Processing System

#function to process student marks
def process_result(marks_list):
    #variable to store total marks
    total = 0
    #loop to calculate total marks
    for mark in marks_list:
        total = total + mark
    #calculate average marks
    average = total / len(marks_list)
    #check pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
#list of student marks
marks = [45, 60, 55, 40, 70]
#call the function and store the result
result = process_result(marks)
#display the result
print("Result:", result)
```

```python
#problem 4: Cab Fare Estimator with Retry Option

#function to calculate cab fare
def calculate_fare(distance, peak_hour):
    base_fare = 50
    rate_per_km = 12
    #calculate normal fare
    fare = base_fare + (distance * rate_per_km)
    #add 25% extra if it is peak hour
    if peak_hour == "yes":
        fare = fare + (fare * 0.25)
    return fare

#allow user to retry using while loop
while True:
    #take distance input from user
    distance = float(input("Enter distance travelled (in km): "))
    #ask if it is peak hour
    peak = input("Is it peak hour? (yes/no): ").lower()
    #call the function to calculate fare
    total_fare = calculate_fare(distance, peak)
    #display the total fare
    print("Total Cab Fare: ₹", int(total_fare))
    #ask user if they want to calculate again
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

```python
#problem 5: Employee Attendance Eligibility Checker

#function to check employee attendance eligibility
def check_attendance(attendance_list):
    present_days = 0
    #loop through the attendance list
    for day in attendance_list:
        if day == "P":
            present_days += 1
    #calculate total number of days
    total_days = len(attendance_list)
    #calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    #check eligibility
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

#example attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
#call the function
result = check_attendance(attendance)
# Display the result
print("Attendance Status:", result)
```

```python
#problem 6: Password Strength Checker

#function to check password strength
def check_password_strength(password):
    #check minimum length
    if len(password) < 8:
        return "Weak Password"

    has_digit = False          #to check if password has a number
    has_special = False        #to check if password has special character
    special_chars = "@#$"      #allowed special characters

    #loop through each character in the password
    for ch in password:
        #check if character is a digit
        if ch.isdigit():
            has_digit = True
        #check if character is a special character
        if ch in special_chars:
            has_special = True
    #final check for password strength
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

#take password input from user
user_password = input("Enter your password: ")
#call the function
result = check_password_strength(user_password)
#display the result
print("Password Strength:", result)
```
